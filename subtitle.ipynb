```python
# we have to make small chunks of audio file to generate subtitle for each line of audio
# we will use pydub library to split the audio file
# read the audio file
x = AudioSegment.from_mp3("news.mp3")
chunks = split_on_silence(x, min_silence_len=500, silence_thresh=-30, keep_silence=500)
# we will save each chunk of audio in a separate file 
# create a directory to store the audio chunks according to its filename

for i, chunk in enumerate(chunks):
    # store in a folder named "chunks"
    chunk.export(f"chunks/chunk{i}.mp3", format="mp3")
```

```python
# count the number of chunks from the folder and play them by ipython
# count the number of chunks
path = "chunks"
files = os.listdir(path)
print(f"Total number of chunks: {len(files)}")

# print the chunks filename
for i in range(len(files)):
    print(files[i])
    
```

```python
# get the timecodes of each chunk and store in a list
# it will be like 00:00:41,333 --> 00:00:42,500

timecodes = []
for i in range(len(files)):
    # get the timecodes of each chunk
    # get the start time of each chunk
    start_time = str(chunks[i].start_seconds)
    # get the end time of each chunk
    end_time = str(chunks[i].end_seconds)
    # append the timecodes in a list
    timecodes.append(start_time + " --> " + end_time)
    
# print the timecodes
for i in range(len(timecodes)):
    print(timecodes[i])
    
```
